# IMPORTS

```python
import os
import pydicom
import numpy as np
from PIL import Image, ImageOps
```

# RESIZE IMAGES FUNCTION

```python
def dcm_to_png(input_path, output_path):
    ds = pydicom.dcmread(input_path)
    pixel_array = ds.pixel_array
    
    print("Valor mínimo de píxel:", np.min(pixel_array))
    print("Valor máximo de píxel:", np.max(pixel_array))
    
    image = Image.fromarray(pixel_array)

    try:
        image.save(output_path)
    except Exception as e:
        print("Error al guardar la imagen:", e)
        return
    
    image_saved = Image.open(output_path)
    
    desired_size = 512
    old_size = image_saved.size
    ratio = float(desired_size)/max(old_size)
    new_size = tuple([int(x*ratio) for x in old_size])
    image_resized = image_saved.resize(new_size, Image.BICUBIC)
    
    delta_w = desired_size - new_size[0]
    delta_h = desired_size - new_size[1]
    padding = (delta_w//2, delta_h//2, delta_w-(delta_w//2), delta_h-(delta_h//2))
    
    image_padded = ImageOps.expand(image_resized, padding, fill='black')
    
    try:
        image_padded.save(output_path)
    except Exception as e:
        print("Error al guardar la imagen redimensionada:", e)
```

# RESIZING TRAIN IMAGES

```python
input_dir = 'train'
output_dir = 'train_png'

for filename in os.listdir(input_dir):
    if filename.endswith('.dcm'):
        input_path = os.path.join(input_dir, filename)
        output_path = os.path.join(output_dir, filename[:-4] + '.png')
        dcm_to_png(input_path, output_path)
```

# RESIZING TEST IMAGES

```python
input_dir = 'test'
output_dir = 'test_png'

for filename in os.listdir(input_dir):
    if filename.endswith('.dcm'):
        input_path = os.path.join(input_dir, filename)
        output_path = os.path.join(output_dir, filename[:-4] + '.png')
        dcm_to_png(input_path, output_path)
```
